```python
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
import json
import pandas as pd
from datetime import datetime, date
import numpy as np

```

```python
DEVELOPER_KEY = ''
YOUTUBE_API_SERVICE_NAME = 'youtube'
YOUTUBE_API_VERSION = 'v3'
```

### Search by Video Name

```python
query = 'THAT GUY - PATAWAWA'

def youtube_video_search(query,querytype='Search'):
    ### querytype: Search, ID , URL
    ### package required: googleapiclient, json, pandas
    
    def youtube_video_state(idyt,url=False):
        youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION,
        developerKey=DEVELOPER_KEY)

        if url == True:
            idyt = idyt.split('/watch?v=')[1]

        # Call the videos.list method to retrieve location details for each video.
        video_response = youtube.videos().list(id=idyt,part='statistics, snippet, status').execute()
        video_result = video_response.get('items', [])[0]
        dictoutput = dict()

        try:
            dictoutput['commentCount'] = video_result['statistics']['commentCount']
            dictoutput['viewCount'] = video_result['statistics']['viewCount']
            dictoutput['dislikeCount'] = video_result['statistics']['dislikeCount']
            dictoutput['likeCount'] = video_result['statistics']['likeCount']
            dictoutput['favoriteCount'] = video_result['statistics']['favoriteCount']
            dictoutput['channelID'] = video_result['snippet']['channelId']
            dictoutput['channelURL'] = 'https://www.youtube.com/channel/'+video_result['snippet']['channelId']
            dictoutput['channeltitle'] = video_result['snippet']['channelTitle']
            dictoutput['videotitle'] = video_result['snippet']['title']
            dictoutput['videodescription'] = video_result['snippet']['description']
            dictoutput['channeltitle'] = video_result['snippet']['channelTitle']
            dictoutput['tags'] = 'Unknown'
            dictoutput['license'] = video_result['status']['license']
            dictoutput['publishedAt'] = video_result['snippet']['publishedAt']
            dictoutput['datetime'] = str(date.today())
        except Exception as inst:
            dictoutput['commentCount'] = 0
            dictoutput['viewCount'] = 0
            dictoutput['dislikeCount'] = 0
            dictoutput['likeCount'] = 0
            dictoutput['favoriteCount'] = 0
            dictoutput['channelID'] = 'Unknown'
            dictoutput['channelURL'] = 'Unknown'
            dictoutput['channeltitle'] = 'Unknown'
            dictoutput['tags'] = 'Unknown'
            dictoutput['license'] = 'Unknown'
            dictoutput['videotitle'] = 'Unknown'
            dictoutput['videodescription'] = 'Unknown'
            dictoutput['publishedAt'] = str(date.today())
            dictoutput['datetime'] = str(date.today())

        return(dictoutput)

    if querytype == 'Search':
        options = {
          'q': query,
          'max_results': 50,
          'type': 'video',
          'order': 'viewCount',
          'videoDuration': 'short',
          'pageToken': None,
          'topicID': '/m/04rlf',
          'location': "52.36800, 4.90360",
          'locationradius': "1000km"}

        youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION,
            developerKey=DEVELOPER_KEY, cache_discovery=False)

          # Call the search.list method to retrieve results matching the specified
          # query term.
        search_response = youtube.search().list(
          q=options['q'],
          part='id,snippet',
          maxResults=options['max_results'],
          type=options['type'],
          order=options['order'],
          pageToken=options['pageToken'],
          videoDuration=options['videoDuration'],  
        #   location=options['location'],
        #   locationRadius=options['locationradius']
        ).execute()
        videos = []
        videosname = []
        url = []
        description = []
        publisheddate = []
        publishedtime = []
        videoID = []
        channelTitle = []

        for search_result in search_response.get('items', []):
            if search_result['id']['kind'] == 'youtube#video':
                videosname.append(search_result['snippet']['title'])
                url.append('https://www.youtube.com/watch?v='+search_result['id']['videoId'])
                videoID.append(search_result['id']['videoId'])
                channelTitle.append(search_result['snippet']['channelTitle'])
                description.append(search_result['snippet']['description'])
                publisheddate.append(search_result['snippet']['publishedAt'][0:10])
                publishedtime.append(search_result['snippet']['publishedAt'][11:19])

        data = pd.DataFrame()
        data['channeltitle'] = channelTitle
        data['videosname'] = videosname
        data['url'] = url
        data['videoID'] = videoID
        data['description'] = description
        data['publisheddate'] = publisheddate
        data['publishedtime'] = publishedtime

        overalldict = dict()
        for videoid_row in videoID[0:5]:
            overalldict[videoid_row] = youtube_video_state(videoid_row,url=False)

        return(overalldict)
    elif querytype == 'ID':
        return(youtube_video_state(query))
    elif querytype == 'URL':
        return(youtube_video_state(query,url=True))
    else:
        return({'Unknown':'Unable to comply'})

youtube_video_search(query)
```
